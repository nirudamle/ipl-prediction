```python
#removing unwanted columns
columns_to_remove=['mid','venue','batsman','bowler','striker','non-striker']
print("Before removing unwanted columns:",df.shape)
df.drop(labels=columns_to_remove,axis=1,inplace=True)
print("After removing unwanted columns:",df.shape)
```

```python
#keeping only consistent teams
df['bowl_team'].value_counts()
```

```python
#keeping only consistent teams
df['bowl_team'].value_counts()
```

```python
consistent_teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals','Mumbai Indians','Kings XI Punjab',
    'Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']

print("Before Removing inconsistent teams:",df.shape)

df=df[(df['bat_team'].isin(consistent_teams))&(df['bowl_team'].isin(consistent_teams))]
print("After Removing inconsistent teams:",df.shape)
```

```python
#Removing the first 5 overs data in every match
df['overs'].unique()
```

```python
print('Before Removing 5 overs teams:',df.shape)
df=df[df['overs']>=5.0]
print('After Removing first 5 overs teams:',df.shape)
```

```python
#converting date object to datetime object
from datetime import datetime
print("Before converting 'date; column from string to datetime:",type(df.iloc[0,0]))

df['date']=df['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d'))
print("After converting 'date; column from string to datetime:",type(df.iloc[0,0]))
```

```python
df.dtypes
```

```python
# df=pd.get_dummies(data=df,columns=['bat_team','bowl_team'],dtype='int')
df = pd.get_dummies(df, columns=['bat_team','bowl_team'], dtype=int)
df
```

```python
df=df[['date','bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils','bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
    'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals','bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
    'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils','bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
    'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals','bowl_team_Royal Challengers Bangalore',
    'bowl_team_Sunrisers Hyderabad','runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5','total']]
```

```python
df.columns
```

```python
#labelling the data
print(df['date'].min())
print(df['date'].max())
```

```python
#class 16th
x=df.drop('total',axis=1)

# splitting the data for training and testing
x_train=x[x['date'].dt.year<=2016] #training inputs(80%)
x_test=x[x['date'].dt.year>=2017] #testing the input(20%)

# After the training and testing data is splitted using date column. As date column is not necessary for modelling it is dropped

x_train.drop('date',axis=1,inplace=True)
x_test.drop('date',axis=1,inplace=True)

# Output Training and Testing
y_train=df[df['date'].dt.year<=2016]['total'].values
y_test=df[df['date'].dt.year>=2017]['total'].values
```

```python
from sklearn.linear_model import LinearRegression
#create object for the class
lr=LinearRegression()
#training the model
lr.fit(x_train,y_train)
```

```python
#model testing
lr_pred=lr.predict(x_test)
lr_pred
```

```python
#Model Evaluation
print("Score of LR on Training Set: ", lr.score(x_train,y_train))
print("Score of LR on Testing Set: ", lr.score(x_test,y_test))
```

```python
#model testing
lr_pred=lr.predict(x_test)
lr_pred
```

```python
# Error Metrics
import numpy as np
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse
print("---------------------------Linear Regression---------------------------")
print("Mean Absolute Error(MAE):",mae(y_test,lr_pred))
print("Mean Squared Error(MSE):",mse(y_test,lr_pred))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,lr_pred)))
```

```python
#How to understand this metics
#mse is having more value , so neglect it
#mae and rmse are close so these are fine 
#because of this we can use as final model
```

```python
#class 17th
# Create  a function for prediction
def predict_score(bat_team='Mumbai Indians',bowl_team='Delhi Daredevils',runs=120,wickets=4,overs=6.2,runs_last_5=33,wickets_last_5=1):
    temp_array=list()
    
    if bat_team=='Chennai Super Kings':
        temp_array=temp_array+[1,0,0,0,0,0,0,0]
    elif bat_team=='Delhi Daredevils':
        temp_array=temp_array+[0,1,0,0,0,0,0,0]
    elif bat_team=='Kings XI Punjab':
        temp_array=temp_array+[0,0,1,0,0,0,0,0]
    elif bat_team=='Kolkata Knight Riders':
        temp_array=temp_array+[0,0,0,1,0,0,0,0]
    elif bat_team=='Mumbai Indians':
        temp_array=temp_array+[0,0,0,0,1,0,0,0]
    elif bat_team=='Rajasthan Royals':
        temp_array=temp_array+[0,0,0,0,0,1,0,0]
    elif bat_team=='Royal Challengers Bangalore':
        temp_array=temp_array+[0,0,0,0,0,0,1,0]
    elif bat_team=='Sunrisers Hyderabad':
        temp_array=temp_array+[0,0,0,0,0,0,0,1]

    if bowl_team=='Chennai Super Kings':
        temp_array=temp_array+[1,0,0,0,0,0,0,0]
    elif bowl_team=='Delhi Daredevils':
        temp_array=temp_array+[0,1,0,0,0,0,0,0]
    elif bowl_team=='Kings XI Punjab':
        temp_array=temp_array+[0,0,1,0,0,0,0,0]
    elif bowl_team=='Kolkata Knight Riders':
        temp_array=temp_array+[0,0,0,1,0,0,0,0]
    elif bowl_team=='Mumbai Indians':
        temp_array=temp_array+[0,0,0,0,1,0,0,0]
    elif bowl_team=='Rajasthan Royals':
        temp_array=temp_array+[0,0,0,0,0,1,0,0]
    elif bowl_team=='Royal Challengers Bangalore':
        temp_array=temp_array+[0,0,0,0,0,0,1,0]
    elif bowl_team=='Sunrisers Hyderabad':
        temp_array=temp_array+[0,0,0,0,0,0,0,1]


    temp_array=temp_array+[runs,wickets,overs,runs_last_5,wickets_last_5]

    #converting to numpy array
    # temp_array=np.array(temp_array)
    # #prediction
    # return int(lr.predict(temp_array))
    temp_array = np.array([temp_array])
    return int(lr.predict(temp_array)[0])
```

```python
final_score=predict_score(bat_team='Mumbai Indians', bowl_team='Royal Challengers Bangalore',runs=100,wickets=0,overs=10.1,runs_last_5=40,wickets_last_5=0)
print('Final Predicted Scoreis {} to {}'.format(final_score-5,final_score+10))
```
